```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
 
calefornia=fetch_california_housing()
df=pd.DataFrame(calefornia.data,columns=calefornia.feature_names)
```

```python
df.head()
```

```python
location_features=df[['Latitude','Longitude']]
housing_features=df[['AveRooms','AveOccup']]
```

```python
from sklearn.cluster import KMeans
 
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(location_features)
df['Cluster']=kmeans.labels_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'], cmap='viridis', s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('K-Means Clusters by Location')
plt.colorbar(label='Cluster')
plt.show()
```

```python
new_point = [[34.2, -119.3]]  # Latitude, Longitude

cluster_id = kmeans.predict(new_point)
"New data belongs to cluster:", cluster_id[0]
```
